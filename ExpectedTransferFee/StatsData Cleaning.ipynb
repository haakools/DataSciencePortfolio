```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading webscraped data

```python
# Getting path to data
workdir = os.getcwd()
data_path = os.path.join(workdir, "skill_data")

# Initalizing new combined dataframe
df = pd.DataFrame()

# Combining all the data and assigining new column as league
for file in os.listdir(data_path):
    print("Loading file " + os.path.join(data_path, file))
    temp_df = pd.read_csv(os.path.join(data_path,file))
    print(temp_df.shape)
    temp_df["league"] = file[:-4].replace("-", " ")
    df = pd.concat([df, temp_df])

# Assigning the NaN entries to be equal to 0. If a problem arises due to setting it equal to zero, revert back here and tweak.
df.fillna(0)
```

### Data Cleaning

```python
df = df.drop(columns="Unnamed: 0")
df = df.reset_index().drop(columns="index")

# Cleaning age column
df["age"] = df["age"].apply(lambda x : str(x)[:2])

# Combining each observation for a year into a single row
stat_feat = [col for col in df.columns]
grouped_df = df.groupby(['player','season'])[stat_feat].first() # Combining all the season data
grouped_df = grouped_df.swaplevel().reset_index('player', drop=True) # Resetting index

# Create start and end year of season
grouped_df['year'] = grouped_df['season'].apply(lambda x : x[5:]).astype('int64')

# Filling nans for GK stats and like
grouped_df = grouped_df.fillna(0)
```

# Exporting to a single csv file

```python
grouped_df.to_csv("stats_data.csv")
```
