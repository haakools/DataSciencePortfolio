```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tqdm
```

# Reading in the dataframes

```python
# Player stats data
df_stats = pd.read_csv("stats_data.csv")

# Player transfer data
df_transfer = pd.read_csv("transfer_data.csv")
df_transfer = df_transfer.drop(columns="Unnamed: 0")
```

```python
# Taking out data needed for the model
df_y = df_transfer[["club_name", "position","player_name", "age", "fee_infladj", "league_name", "year"]]
df_y = df_y[df_y["year"]>2017] # getting 2018 and onwards


# Getting all the unique players
player_names = df_y["player_name"].unique()
print("Players with transfers since 2016\n", player_names, "\nAmount: ",len(df_y["player_name"].unique()) )
```

### Merging datasets

```python

# Renaming column to combine on
df_stats = df_stats.rename(columns={'player':'player_name'})

# Removing duplicate columns before merging
df_stats = df_stats.drop(columns=['position', 'age', 'nationality', 'season', 'squad', 'league'] )

```

```python
features = []
for col in df_stats.columns:
    if df_stats[col].dtype == 'float64':
        features.append(col)
        
features.remove('birth_year')
print(features)
```

```python
df = pd.merge(df_stats, df_y, how='outer', on=['player_name', 'year'])
```

```python
# Filling seasons with missing data
df[features] = df[features].fillna(0)
```

### Defining weighted mean function

```python
def weighted_mean(data):
    n = len(data)
    w = []
    for i in reversed(range(n)):
        w.append(np.exp(-i))
    w = np.array(w)
    
    return np.sum(data*w)/np.sum(w)
```

### Applying weighted mean function on all season before a transfer

```python
from datetime import datetime
```

```python
features
```

```python
X = []
y = []

timer1 = []
timer2 = []
timer3 = []
for player in tqdm.tqdm(np.flip(player_names)):
    
    # Get player name
    player_df = df[df["player_name"] == player]
    
    # Sort by years
    player_df = player_df.sort_values("year")
    
    # Filling in birth year for missing intervals
    player_age = player_df["birth_year"].mean()
    player_df["birth_year"].fillna(player_age, inplace=True)

    # Calculating age for every player
    player_df["age"] = player_df["year"]-player_age

    # Get list of years where transfer occur
    transfer_years = player_df[~player_df.fee_infladj.isna()]["year"].to_numpy()
    

    
    # Loop over transfers to squish previous stats
    for year in transfer_years:
        
        time1 = datetime.now()
        
        # Years before transfer
        temp_df = player_df[player_df["year"]<=year]
        timer1.append(datetime.now()-time1)
                
        time2 = datetime.now()
        # Weighted mean
        X.append(temp_df[features].apply(weighted_mean, axis=0).to_numpy())
        timer2.append(datetime.now()-time2)
        
        time3 = datetime.now()
        y.append(player_df[df_y.columns][player_df["year"] == year].to_numpy()[0])
        timer3.append(datetime.now()-time3)

        
    
```

```python
X_df = pd.DataFrame(X)
X_df.columns = features
Y_df = pd.DataFrame(y)
Y_df.columns = df_y.columns
```

### Scaling the data

```python
# Adjust transferfee for age to see hidden skill correlation
def gauss(x):
    a, x0, sigma = 3.11, 25, 5.85
    return a * np.exp(-(x - x0)**2 / (2 * sigma**2))
#X_df["age_fee"] = Y_df["fee_infladj"]*Y_df["age"].apply(lambda x: gauss(x))
Y_df["age_fee"] = Y_df["fee_infladj"]*Y_df["age"].apply(lambda x: gauss(x))

# Dropping nan in age
X_df = X_df[~Y_df["age"].isna()]
Y_df = Y_df[~Y_df["age"].isna()]
```
