# Beautiful soup intro

### downloading pages

Request EPL homepage

```python
import requests
page = requests.get("https://www.transfermarkt.com/premier-league/startseite/wettbewerb/GB1/saison_id/2018", headers={'User-Agent': 'Custom'})
page
```

View content of response

```python
page.content
```

### parsing pages

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
```

Pretty printing

```python
print(soup.prettify())
```

### parsing tables

Find table of clubs

```python
table = soup.find(class_="grid-view", id="yw1")
table
```

Parse table

```python
table_body = table.find('tbody')

data = []

rows = table_body.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data.append([ele for ele in cols if ele]) 
```

```python
data
```

### combining data in dataframes

```python
import pandas as pd

data = pd.DataFrame(data)
data = data[data.columns[0:6]] 
data.columns = ['club', 'club_abr', 'squad', 'age', 'foreign_players', 'total_mkt_value']
data
```
